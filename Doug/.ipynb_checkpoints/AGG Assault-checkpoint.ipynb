```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
atlanta_crime_data = "./COBRA-2009-2018.csv"


# Read the Mouse and Drug Data and the Clinical Trial Data
atlanta_crime_df = pd.read_csv(atlanta_crime_data)

atlanta_crime_df.head()
```

```python
atlanta_crime_2018_df = atlanta_crime_df.loc[atlanta_crime_df['Report Date'].str.contains('2018')]
#df_grp = df.groupby(['rank', 'discipline'])
agg_assault_2018 = atlanta_crime_2018_df.loc[atlanta_crime_2018_df['UCR Literal']=='AGG ASSAULT']
#atlanta_crime_2018_grp = atlanta_crime_2018_df.groupby(['Neighborhood','UCR Literal'])
agg_assault_2018 = agg_assault_2018.groupby('Neighborhood')
agg_assault_crime = agg_assault_2018['UCR Literal'].count()

agg_assault_2018 = pd.DataFrame({'AGG Assault Crime 2018': agg_assault_crime})
agg_assault_2018.head()
```

```python
agg_assault_2018_graph = agg_assault_2018.sort_values("AGG Assault Crime 2018", axis=0, ascending=False).head(10)
```

```python
agg_assault_2018_graph
```

```python
#robbery_pedestrian_2018_graph.plot(kind='bar',x='Neighborhood',y='Pedestrian Crime Count 2018')
agg_assault_2018_graph.plot.bar(figsize=(15, 8), color='red')

plt.xlabel("Neighborhood", labelpad=14, fontsize=18)
plt.ylabel("Crimes Reported", labelpad=14, fontsize=18)
plt.title("Agg Assault Crime Count by Neighborhood 2018", y=1.02, fontsize=23)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), fontsize=18)
plt.tick_params(labelsize=14)
# the plot gets saved to 'output.png'
plt.savefig('Agg Assault Crime Count by Neighborhood 2018.png', bbox_inches = "tight")
```
